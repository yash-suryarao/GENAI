a) Chain-of-Thought Prompting (COT)

```python
# COT usinf Gemini

import google.generativeai as genai

genai.configure(api_key=gemini_api_key)

model = genai.GenerativeModel("gemini-1.5-flash")

prompt = """
Think step-by-step about India's significance on the global stage.
First, consider its geographical size and strategic location.
Then, evaluate its rich cultural history and influence on world traditions.
Next, analyze India's growing economy and technological advancements.
Finally, explain how its political influence impacts global diplomacy. Please give a reasoned response.
"""

response = model.generate_content(prompt)

print(response.text)
```

```python
# COT # OpenAI using Open router
from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=openrouter_api_key,
)

completion = client.chat.completions.create(
  model="openai/gpt-oss-20b:free",
  messages=[
    {
      "role": "user",
      "content": """
        Think step-by-step about India's significance on the global stage.
        First, consider its geographical size and strategic location.
        Then, evaluate its rich cultural history and influence on world traditions.
        Next, analyze India's growing economy and technological advancements.
        Finally, explain how its political influence impacts global diplomacy. Please give a reasoned response.
      """
    }
  ]
)
print(completion.choices[0].message.content)
```

```python
# COT Hugging face
from transformers import pipeline

genrator = pipeline('text-generation', model='gpt2')

prompt = """
Think step-by-step about India's significance on the global stage.
First, consider its geographical size and strategic location.
Then, evaluate its rich cultural history and influence on world traditions.
Next, analyze India's growing economy and technological advancements.
Finally, explain how its political influence impacts global diplomacy. Please give a reasoned response.
"""

output = genrator(prompt, max_length=50, num_return_sequences=1)

print("\nResponse: ",output[0]['generated_text'])
```

```python
# Role-based Instruction OpenAI using Open Router
from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=openrouter_api_key,
)

completion = client.chat.completions.create(
  model="openai/gpt-oss-20b:free",
  messages=[
    {
      "role": "user",
      "content": """
        Summarize India briefly.
        Focus on specific key aspects (geography, culture, history, economy).
        Keep the response concise (3-4 sentences).
      """
    }
  ]
)
print(completion.choices[0].message.content)
```
